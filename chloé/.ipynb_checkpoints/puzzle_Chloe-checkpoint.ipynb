## Puzzle méthode 1

Ce puzzle est paru en 1994 dans un magazine de supplément offert avec le *Weekend Australian*. 

>Dan est en charge de la promenade de 5 chiens, qui lui ont été confiés par 5 femmes. Seulement, il ne se rappelle plus quel chien appartient à qui, et de plus, il mélange les prénoms des propriétaires. Voici ce dont il est certain.
>
>Le chien de Mme Brown, Loopsie ou Mooksie, n'est pas un terrier. Woopsie, le caniche, n'appartient pas à Mary. Le setter de Marion n'est pas Loopsie. Le basset de Margie s'appelle Smooksie ou Mooksie. Hilary, qui n'est pas Mme Grey, est la maîtresse de Poopsie ou Smooksie. Martha Black n'est pas la maîtresse du dalmatien. Poopsie n'est pas à Mme Grey, mais Mooksie est à Mme White, dont le prénom n'est pas Marion.
>
>Question : sachant que Mme Green est l'une des propriétaires de chien, trouver son prénom, le nom de son chien, et la race de son chien.
>
>Indication : posez-vous d'abord la même question pour Mme Brown, puis Mme White, puis Mme Grey.

$A\implies B$ est la même chose que ($B$ ou $non(A)$)

$A \iff B$ est la même chose que $A$ == $B$

```python
prenoms= ["Mary", "Marion", "Margie", "Hilary", "Martha"]
noms= ["Brown", "Grey", "Black", "White", "Green"]
races= ["terrier", "caniche", "dalmatien", "setter", "basset"]
chiens= ["Loopsie", "Mooksie", "Smooksie", "Poopsie", "Woopsie"]
```

L'ensemble de toutes les combinaisons possibles est:

```python
from itertools import product
comb= [ [p, n, r, c] for p,n,r,c in product(prenoms, noms, races, chiens)]
```

Commençons à trier à partir des informations données par Dan 

$n=$ Brown $\implies c=$ Loopsie ou Mooksie (le chien de Mme Brown est Loopsie ou Mooksie)

```python
filtre1= [ [p,n,r,c] for p,n,r,c in comb if (c in ["Loopsie", "Mooksie"]) or (not n=="Brown")   ]
```

$n=$ brown $\implies r\ne$ terrier (le chien de Mme Brown n'est pas un caniche)

```python
filtre2= [ [p,n,r,c] for p,n,r,c in filtre1 if (r!= "terrier") or (not n=="Brown")]
```

$c$ = Woopsie $\iff$ $r$ = caniche (Woopsie est un caniche)


```python
filtre3= [ [p,n,r,c] for p,n,r,c in filtre2 if (c == "Woopsie") == (r == "caniche")]
```

$c=$ Woopsie $\implies p\ne$ Mary (Woopsie n'appartient pas à Mary)

```python
filtre4= [ [p,n,r,c] for p,n,r,c in filtre3 if  p != "Mary" or not(c=="Woopsie")]
```

$p$ = Marion $\iff r$ = setter (Marion a un setter)

```python
filtre5= [ [p,n,r,c] for p,n,r,c in filtre4 if  (r=="setter") == (p=="Marion")]
```

$p=$ Marion $\implies c\ne$ Loopsie (le setter de Marion n'est pas Loopsie)

```python
filtre6= [ [p,n,r,c] for p,n,r,c in filtre5 if  c!="Loopsie" or not(p=="Marion")]
```

$p$ = Margie $\iff r$ = basset (Margie a un basset)

```python
filtre7= [ [p,n,r,c] for p,n,r,c in filtre6 if  (r=="basset") == (p=="Margie")]
```

$r=$ basset $\implies c=$ Smooksie ou Mooksie (Ce basset s'appelle Smooksie ou Mooksie)

```python
filtre8= [ [p,n,r,c] for p,n,r,c in filtre7 if  (c in ["Smooksie", "Mooksie"]) or not(r=="basset")  ]
```

$p=$ Hilary $\implies n\ne$ Grey (Hilary n'est pas Mme Grey)

```python
filtre9= [ [p,n,r,c] for p,n,r,c in filtre8 if  not(n=="Grey") or not(p=="Hilary")   ]
```

$p=$ Hilary $\implies c=$ Poopsie ou Smooksie (Hilary est la maîtresse de Poopsie ou Smooksie)

```python
filtre10=[ [p,n,r,c] for p,n,r,c in filtre9 if  (c in ["Poopsie", "Smooksie"]) or not(p=="Hilary")   ]
```

$p$ = Martha $\iff n$ = Black (Martha est Mme Black)

```python
filtre11=[ [p,n,r,c] for p,n,r,c in filtre10 if  (p=="Martha") == (n=="Black")]
```

$p=$ Martha $\implies r\ne$ dalmatien (Martha n'est pas la maîtresse du dalmatien)

```python
filtre12= [ [p,n,r,c] for p,n,r,c in filtre11 if  not(r=="dalmatien") or not(p=="Martha")   ]
```

$c=$ Poopsie $\implies n\ne$ Grey (Poopsie n'est pas à Mme Grey)

```python
filtre13= [ [p,n,r,c] for p,n,r,c in filtre12 if not(n=="Grey") or not(c=="Poopsie")]
```

$c$ = Mooksie $\iff n$ = White (Mooksie est à Mme White)

```python
filtre14= [ [p,n,r,c] for p,n,r,c in filtre13 if (c=="Mooksie") == (n=="White")   ]
```

$n=$ White $\implies p\ne$ Marion (le prénom de Mme White n'est pas Marion)

```python
filtre15= [ [p,n,r,c] for p,n,r,c in filtre14 if  not(p=="Marion") or not(n=="White")]
```

```python
print(len(filtre15))
```

Plus que 27 combinaisons possibles, sur les 625 de départ!
Voyons toutes celles où le nom de famille est Green:

```python
S = filtre15
[ [p,n,r,c] for p,n,r,c in S if n=="Green"]
```

On ne peut pas conclure sans travailler un peu plus. Mais pour Mme Brown:

```python
[ [p,n,r,c] for p,n,r,c in S if n=="Brown"]
```

On voit que Mary Brown possède un dalmatien qui s'appelle Loopsie (deux nouvelles conditions).

$r=$ dalmatien $\iff$ $c=$ Loopsie

$n=$ Brown $\iff$ $p=$ Mary

```python
filtre16 = [ [p,n,r,c] for p,n,r,c in filtre15 if (r=="dalmatien") == (c=="Loopsie")]
filtre17 = [ [p,n,r,c] for p,n,r,c in filtre16 if (n=="Brown") == (p=="Mary")]
```

On voit que Mme White est Margie et que le basset s'appelle Smooksie:

$n=$ White $\iff$ $p=$ Margie

```python
filtre18 = [ [p,n,r,c] for p,n,r,c in filtre17 if (n=="White") == (p=="Margie")]
filtre19 = [ [p,n,r,c] for p,n,r,c in filtre18 if (r=="basset") == (c=="Mooksie")]
```

## Puzzle méthode 2

On crée deux fonctions anonymes et jetables (lambda : http://www.polora.fr/atelier-python/fr/section3/fonction-lambda), une qui gère la relation implique, l'autre qui gère la relation equivalent.

Pour créer nos filtres dans la méthode 1, nous n'avons eu besoin que ces deux relations.

```python
implique=  lambda A, B : B or not A   # prend A et B en paramètres et renvoie True si B or not A est vrai
equivalent=lambda A, B : A == B       # même principe
```

```python
filtre= [ [p,n,r,c] for p,n,r,c in comb if all([
    implique(n=="Brown", c in ["Loopsie", "Mooksie"]),
    implique(n=="Brown", r!= "terrier"),
    equivalent(c == "Woopsie", r == "caniche"),
    implique(c=="Woopsie", p != "Mary"),
    equivalent(r=="setter", p=="Marion"),
    implique(p=="Marion", c!="Loopsie"),
    equivalent(r=="basset", p=="Margie"),
    implique(r=="basset", c in ["Smooksie", "Mooksie"]),
    implique(n=="Grey", not(p=="Hilary")),
    implique(p=="Hilary", c in ["Poopsie", "Smooksie"]),
    equivalent(p=="Martha", n=="Black"),
    implique(p=="Martha", not(r=="dalmatien")),
    implique(n=="Grey", not(c=="Poopsie")),
    equivalent(c=="Mooksie", n=="White"),
    implique(n=="White", not(p=="Marion")) ]) # fin du all([...])
    ] # fin du [ de départ
```

```python
print(len(filtre))
```

```python
for element in filtre:
    print(element)
```

On retrouve bien nos 15 filtres précédents et, au final, une liste de possibilités réduite à 27 associations. C'est juste une façon plus rapide et réduite d'écrire nos filtres.

C'est pour les derniers filtres qu'une autre méthode (graphique) a été utilisée dans la correction de cet exercice.

```python
import networkx as nx

# on crée un nouveau graphe avec 27 noeuds, autant que de solutions restantes dans notre puzzle
G= nx.Graph()
G.add_nodes_from(range(27))

# la fonction suivante compare deux éléments de la liste qui s'affiche juste au-dessus (la liste des solutions restantes une fois les
# filtres appliqués), l'élément à l'index i et l'élément à l'index j.
# si ces deux éléments ont au moins un point commun (même nom, prénom, nom du chien, race), la fonction renvoie False sinon elle renvoie True
def compatible(i,j):
    L= filtre[i]
    LL= filtre[j]
    if L[0]==LL[0] or L[1]==LL[1] or L[2]==LL[2] or L[3]==LL[3]:
        return False
    else:
        return True
        
# on applique cette fonction en comparant le premier élément à tous les suivants, puis le second à tous les suivants et ainsi de suite...
# dès que deux éléments sont différents l'un de l'autre, on crée une arête de graphe entre l'index i et l'index j
for i in range(27):
    for j in range(i+1, 27):
        if compatible(i,j):
            G.add_edge(i,j)

nx.draw(G, with_labels=True)
```

Si ça n'est pas suffisamment clair, pour comprendre, on regarde ce que renvoie comme résultat la fonction _compatible_ appliquée :

```python
compteur = 0
for i in range(27):
    for j in range(i+1, 27):
        if compatible(i,j):
            compteur += 1 
            print(compteur)
            print(filtre[i], filtre[j])
```

On voit donc qu'il y a 121 combinaisons possibles de deux éléments complètement différents. Donc 121 arêtes.

Pour trouver la solution à ce problème, la correction propose d'utiliser les fonctions mises à disposition par Networkx.

En se documentant (https://fr.wikipedia.org/wiki/Clique_(th%C3%A9orie_des_graphes), on apprend qu'une clique est un sous-ensemble de sommets dans le graphe qui sont tous adjacents deux à deux.

Puis (https://beauguitte.github.io/analyse-de-reseau-avec-python/C4_partitions.html), on voit que la méthode _find\_cliques()_ de Networkx renvoie le nombre de sommets dans la plus grande clique.
Comme nous avons 5 résultats à trouvers (5 propriétaires, chiens, races), nous cherchons la plus grande clique à 5 sommets (les 5 éléments les plus éloignés donc les plus différents).

Il suffit ensuite d'afficher à quel élément de filtre les résultats obtenus se rapportent

```python
for clique in nx.find_cliques(G):
    if len(clique) == 5:
        print(clique)
```
